### Task 3 - Image to Pencil Sketch with Python

## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt 
```

## Imported Data

```python
img = cv2.imread('dog.jpg')
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(image)
plt.axis('off')
plt.title('Original_image')
plt.show()
```

## Gray Image

```python
ima_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(ima_gray,cmap='gray')
plt.axis('off')
plt.title('Gray Scale Image')
plt.show()
```

## Inverting image 

```python
img_invert = cv2.bitwise_not(ima_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

## Smooting Image

```python
img_smoo = cv2.GaussianBlur(img_invert,(21,21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoo,cmap='gray')
plt.axis('off')
plt.title('Smoothing image')
plt.show()
```

## Pencil Sketch

```python
img_pencil = cv2.divide(ima_gray,255-img_smoo, scale = 255)
plt.figure(figsize=(8,8))
plt.imshow(img_pencil,cmap='gray')
plt.axis('off')
plt.title('Final Pencile Sketch')
plt.show()
```

## Final Image

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image)
plt.axis('off')
plt.title('Orginal Image')

plt.subplot(1,2,2)
plt.imshow(img_pencil,cmap='gray')
plt.axis('off')
plt.title('Final Pencil Sketch')
plt.show()
```
